#### **plt.subplots() & plt.subplot()**
- subplot과 subplots 모두 한번에 여러 그래프를 보여주기 위해 사용되는 코드이지만 사용법이다르다.

```python
# subplot의 예
import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.subplot(2,1,2)
plt.show()
```

```python
# subplots의 예
fig, axes = plt.subplots(nrows=2, ncols=1)
plt.show()
```
즉 subplot는 일일이 다 설정해줘야하지만, subplots는 한번에 설정해 줄 수 있다.

```python
# subplots의 장점 => subplot은 부분을 지시하는 키워드이기에 반환값으로 axes만 받을 수 있다.
ax = plt.subplot(2,1,1)
# subplot은 하나의 값(axes)만 받을 수 있다.
```

#### **np.linspace**
> 인자를 3개를 기본으로 넣어주면 되는데, 구간 시작점, 구간 끝점, 구간 내 숫자 개수 3가지를 순서대로 입력해주면 된다.
- 끝점을 포함하지 않고 싶으면 endpoint = False를 추가해주면 되고,
- retstep = True를 추가하시면 숫자 사이의 간격이 얼마인지에 대한 값을 같이 반환하게 된다.

- 간격이 촘촘할수록 그래프를 그리는 상황 등에서 실제 그래프 개형과 더 가까워진다는 장점이 있지만, 어레이 사이즈가 커져서 다루기 힘들어질 수 있다는 단점이 있다.

- matplotlib.pyplot

````python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(1,2,'ko')
plt.plot([0,2], [0,4], 'r--')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.title('The Graph Title')
````

````python
_, axs = plt.subplots(1,2,figsize = (8,5))
# 1,2 : 하나의 행과 2개의 열 => 2개의 도표가 나란히 있음
# figsize = (너비, 높이)
axs[0].plot(np.random.randn(10,5))
axs[1].imshow(np.random.randn(10,5))
````

````python
_, axs = plt.subplots(1,2,figsize = (8,5))
# 1,2 : 하나의 행과 2개의 열 => 2개의 도표가 나란히 있음
# figsize = (너비, 높이)
axs[0].plot(np.random.randn(5,3))
axs[1].imshow(np.random.randn(5,3))
````

````python
# np.linspace
x = np.linspace(-4,4,42)
y = x**2
plt.plot(x,y, 's-')
````

### Soft Coding
- 변경할 수 있는 parameter에 대한 변수를 만드는 개념

````python
# Sigmoid Function - activation function : non-linear
x = np.linspace(-4,4,42)
alpha = 1.4
beta = 2

num = alpha
den = 1 + np.exp(-beta*x)
fx = num / den
plt.plot(x,fx,'s-')
````
